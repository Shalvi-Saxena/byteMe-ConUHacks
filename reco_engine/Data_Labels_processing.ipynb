```python
import pandas as pd
```

```python
df1 = pd.read_csv("/content/pricerunner_aggregate.csv")
df2 = pd.read_csv("/content/shopmania.csv")
df3 = pd.read_csv("/content/skroutz_aggregate.csv")
```

```python
df1.columns = ['index','Title','x','y','des','z','Label']
```

```python
df2.columns = ['index','Title','x','Label']
```

```python
df3.columns = ['index','Title','x','y','des','z','Label']
```

```python
df1 = df1[['Title','Label']]
df2 = df2[['Title','Label']]
df3 = df3[['Title','Label']]
```

```python
df1.Label.value_counts().keys()
```

```python
df2.Label.value_counts().keys()
```

```python
df3.Label.value_counts().keys()
```

```python
all_categories = {}
```

```python
for i in df1.Label.value_counts().keys():
  print(i+'\n')
  x = int(input("Label"))
  all_categories[i] = x
  print('-'*100)
```

```python
for i in df2.Label.value_counts().keys():
  print(i+'\n')
  x = int(input("Label"))
  all_categories[i] = x
  
  print('-'*100)
```

```python
for i in df3.Label.value_counts().keys():
  print(i+'\n')
  x = int(input("Label"))
  all_categories[i] = x
  
  print('-'*100)
```

```python
import json

with open("progress.json", "w") as fp:
    json.dump(all_categories , fp) 
```

```python
frames = [df1, df2, df3]

df = pd.concat(frames)
```

```python
df.head()
```

```python
df["LabelEncoding"] = df["Label"].map(all_categories)
```

```python
df["LabelEncoding"].value_counts()
```

```python
training_df = df[["Title","LabelEncoding"]]
```

```python
training_df.to_csv("training_df.csv",index=False)
```
